```python
import pandas as pd
df = pd.read_csv('선호도.csv')
```

```python
df

# 1. 어떤 유사도 알고리즘을 사용 할 것 인지?
# 2. 구한 유사도를 어떻게 응용 해줄 것 인지?


# '''
# - 1, 2, 3, 4를 어떻게 표현할지 고민
# 1, 2, 3, 4는 선호도가 증가하는것을 나타내긴 하지만,
# 선호도 4가 선호도 1보다 4배 더 좋아하는것을 나타내는건 아니다.

# - 자카드 유사도
# 가졌다, 안가졌다
# 좋아한다, 안한다 <- 1과 0의 방향일 뿐 
# '''

# '''
# - 구한 유사도를 어떻게 추천으로 응용 해줄것인지?
# A와 B가 유사한데
# A가 좋아하는 식물을 B한테 어떻게 추천 해줄것인지?
# '''
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(df, df)

# 코사인 유사도
# 벡터의 "방향"만 보기 때문에 선호도의 "점수"는 중요하지 않다
```

```python
cosine_sim
```

```python
# 기존의 데이터 형식으로는 식물간의 피어슨 유사도가 나옴 
print(df.corr(method="pearson"))
```

```python
# 데이터 transposer() <- 행렬 전치 (행과 열을 바꿈)
pearson_df = df.transpose()
pearson_df.corr(method='pearson') #피어슨
# 출력 결과 NaN이 나옴 -> 8번과 9번은 [0, 0, 0, 0], [4, 4, 4, 4]
# 어떤것도 좋고, 어떤것도 싫은 애들이니까 NaN이 나옴
```

```python
spearman_df = df.transpose()
spearman_df.corr(method='spearman') # spearman
```

```python
kendall_df = df.transpose()
kendall_df.corr(method='kendall') # kendall
```

```python
# 전치된 행렬 t_df의 8번 유저와 9번 유저 데이터를 바꾸고 pearson유사도 계산
nt_df = df.transpose()
nt_df[8] = [2 ,3, 4, 1]
nt_df[9] = [1, 4, 2, 3]
nt_df.corr(method='pearson')
```
